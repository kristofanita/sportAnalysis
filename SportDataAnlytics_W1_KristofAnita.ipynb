1.	Load the datafile called Football_ENG_2017_18 (this contains 6 variables: the date, home team, away team, goals scored (FTHG), goals against (FTAG) and the result (H- home win, D- draw, A – away win).

```python
from datetime import datetime
import pandas as pd
import numpy as np
#import datetime
football = pd.read_excel("Football_ENG_2017_18.xlsx")
football['HalfSeason'] = -1
for i in range(len(football['Date'])):
  f = football['Date'][i]
  firstHalf = True
  if type(f) == str:
    firstHalf = pd.to_datetime(f) < pd.to_datetime('20180101')
    # print(f)
  else:
    firstHalf = pd.to_datetime(f) < pd.to_datetime('20180101')
  if firstHalf:
    football['HalfSeason'][i] = 0
  else:
    football['HalfSeason'][i] = 1

football.head()
```

2.	Create a value for a home wins (win= 1, draw=0.5, loss= 0) and away wins and a count variable for each game (=1).

```python
football["homeWins"] = np.where(football['FTR'] == 'H', 1, np.where(football['FTR'] == 'D', 0.5, 0))
football["awayWins"] = np.where(football['FTR'] == 'A', 1, np.where(football['FTR'] == 'D', 0.5, 0))
football["count"] = 1
football
```

```python
#create files

#save to excel files, based on conditions
football[football['HalfSeason'] == 0].to_excel('EPL_17.xlsx')
football[football['HalfSeason'] == 1].to_excel('EPL_18.xlsx')


#create the new dataframes and rename columns
EPL17 = pd.read_excel('EPL_17.xlsx')
EPL17Home = EPL17.groupby(['HomeTeam', 'Div'])['HomeTeam', 'Div', 'count', 'homeWins', 'FTHG', 'FTAG'].sum().reset_index()
EPL17Home.rename(columns={'HomeTeam':'Team', 'count':'homeCount', 'FTHG':'homeFTHG', 'FTAG':'homeFTAG'}, inplace=True)

EPL17Away = EPL17.groupby(['AwayTeam', 'Div'])['AwayTeam', 'Div', 'count', 'awayWins', 'FTHG', 'FTAG'].sum().reset_index()
EPL17Away.rename(columns={'AwayTeam':'Team', 'count':'awayCount', 'FTHG':'awayFTHG', 'FTAG':'awayFTAG'}, inplace=True)
EPL17Home
```

4.	Then merge the home and away dataframes.

```python
Season17 = pd.merge(EPL17Home, EPL17Away, on='Team')
Season17.head()
```

5.	Sum the values of home and away wins, games, goals for and goals against, then create the values for win percentage (wpc) and the Pythagorean expectation (pyth).

```python
#First sum the values
Season17['sumOfWins'] = Season17['homeWins'] + Season17['awayWins']
Season17['totalGamesPlayed'] = Season17['homeCount'] + Season17['awayCount']

Season17['totalGoalsFor'] = Season17['homeFTHG'] + Season17['awayFTAG']
Season17['totalGoalsAgainst'] = Season17['homeFTAG'] + Season17['awayFTHG']
#Then count the win percentage and the pyth
Season17['wpc17'] = Season17['sumOfWins'] / Season17['totalGamesPlayed']
Season17['pyth17'] = Season17['totalGoalsFor']**2 / (Season17['totalGoalsFor']**2 + Season17['totalGoalsAgainst']**2)
Season17
```

6.	Now repeat steps 4-6 for the 2018 games. Be sure to give different names for wpc and pyth in 2017 and 2018.

```python
#create the new dataframes for the season 2018 and rename columns
EPL18 = pd.read_excel('EPL_18.xlsx')
EPL18Home = EPL18.groupby(['HomeTeam', 'Div'])['HomeTeam', 'Div', 'count', 'homeWins', 'FTHG', 'FTAG'].sum().reset_index()
EPL18Home.rename(columns={'HomeTeam':'Team', 'count':'homeCount18', 'homeWins':'homeWins18', 'FTHG':'homeFTHG18', 'FTAG':'homeFTAG18'}, inplace=True)

EPL18Away = EPL18.groupby('AwayTeam')['AwayTeam', 'count', 'awayWins', 'FTHG', 'FTAG'].sum().reset_index()
EPL18Away.rename(columns={'AwayTeam':'Team', 'count':'awayCount18', 'awayWins':'awayWins18', 'FTHG':'awayFTHG18', 'FTAG':'awayFTAG18'}, inplace=True)
EPL18Home

#Now merge 2018 Home and Away dfs
Season18 = pd.merge(EPL18Home, EPL18Away, on='Team')
Season18

#sum the values in 2018
Season18['sumOfWins18'] = Season18['homeWins18'] + Season18['awayWins18']
Season18['totalGamesPlayed18'] = Season18['homeCount18'] + Season18['awayCount18']
Season18['totalGoalsFor18'] = Season18['homeFTHG18'] + Season18['awayFTAG18']
Season18['totalGoalsAgainst18'] = Season18['homeFTAG18'] + Season18['awayFTHG18']

#Then count the win percentage and the pyth in 2018
Season18['wpc18'] = Season18['sumOfWins18'] / Season18['totalGamesPlayed18']
Season18['pyth18'] = Season18['totalGoalsFor18']**2 / (Season18['totalGoalsFor18']**2 + Season18['totalGoalsAgainst18']**2)

Season18
```

7.	Now merge 2017 and 2018 summary files.

```python
AllSeason = pd.merge(Season17, Season18, on=['Team'])
AllSeason
```

8.	Now generate a correlation matrix for the wpc and pyth variables for 2017 and 2018

```python
correlationMatrix = AllSeason[['wpc17','wpc18','pyth17','pyth18']].corr()
correlationMatrix
```
